### Exercice: refaire les moyennes + count ci dessus (en essayant de ne pas regarder ;p)

On va virer toutes les lignes qui contiennent moins de 10 ventes, parce que ce n'est pas très représentatif

Par ailleurs, le banquier m'a donné un budget. Tout ce qui est au dessus de 250 000€ est hors limite pour moi. (Bye bye le Touquet...)



Il faut donc trouver un moyen d'exclure ces lignes

<blockquote> Facile, il suffit de faire "Where nb_lines > 10 and vf < 250000 "</blockquote>
    
Hélas non. On va voir ensemble que ce n'est pas si simple, en tout cas pas en SQL

# WHERE

```python
(
    appt_nord
    .groupby("Commune")
    ["valeur_fonciere"]
    .agg(["mean", "count"])
    .reset_index()
    .query("count > 10 and mean < 250000")
)
```

```python
query = """
SELECT Commune,
CAST(MEAN(valeur_fonciere) AS INTEGER) as vf,
COUNT(*) AS nb_lines
FROM appt_nord

GROUP BY Commune

"""
duckdb.sql(query)
```

Ordre des opérations SQL

FROM >> JOIN >> WHERE >> GROUPBY >> HAVING >> SELECT >> ORDER BY

# Subquery / CTE

```python
query = """
SELECT 
FROM (
    SELECT Commune,
    CAST(MEAN(valeur_fonciere) AS INTEGER) as vf,
    COUNT(*) AS nb_lines
    FROM appt_nord
    GROUP BY Commune
)
WHERE nb_lines > 10
"""

duckdb.sql(query)
```

```python
query = """
WITH aggregate_table AS (
    SELECT Commune,
    CAST(MEAN(valeur_fonciere) AS INTEGER) as vf,
    COUNT(*) AS nb_lines
    FROM appt_nord
    GROUP BY Commune
)

SELECT * FROM aggregate_table
WHERE nb_lines > 10
"""
duckdb.sql(query)
```

# 'Having' when 'where' doesn't work

```python
query = """
SELECT Commune,
CAST(MEAN(valeur_fonciere) AS INTEGER) as vf,
COUNT(*) AS nb_lines
FROM appt_nord

GROUP BY Commune

HAVING nb_lines > 10
"""
duckdb.sql(query)
```
